```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df= pd.read_csv('/content/drive/MyDrive/CSCI 5502 Data Mining/01 Project/Data/sample_usable_data_for_initial_analysis.csv')
df.head()
```

1. Find the most popular station?

1. Check the distribution of wait times

```python
data= df.copy()

for i in data['station_id'].unique():
  station_filter=i
  date_filter='2023-10-31'
  data_viz= data[(data['station_id']==station_filter) & (data['date_rnd']==date_filter)]

  wait_times= data_viz.bike_wait_time


  print('Mean',np.mean(wait_times))
  print('SD',np.std(wait_times))
  plt.hist(data_viz.bike_wait_time, density = True, histtype ='bar')
  plt.title(i+" on "+date_filter)
  plt.show()
```

```python
data= df.copy()

date_filter='2023-10-31'
data = data[data['date_rnd']==date_filter]


grouped_df = data.groupby(['time_rnd']).agg({'new_docks_available': 'sum'}).reset_index()
grouped_df= grouped_df[grouped_df['time_rnd']!= '00:00:00'] # Exclude 12AM as the data is high

plt.figure(figsize=(15, 4))
plt.fill_between(grouped_df.time_rnd,grouped_df.new_docks_available)
plt.show()
```
